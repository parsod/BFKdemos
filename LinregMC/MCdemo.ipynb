```python
from numpy import *
from destfunctions import *
import matplotlib.pyplot as plt

plt.rc('font', size=16)

x=array([2.0, 3.0, 4.0, 5.0])
y=array([4.0,3.6,2.9,2.5])
xsig=ones(4)*0.2
ysig=ones(4)*0.1
xrange=array([1.0,6.0])
#plt.xlim(xrange[0],xrange[1])
plt.errorbar(x,y,fmt='bo',xerr=xsig,yerr=ysig,capsize=5)
p=polyfit(x,y,1)
plt.plot(xrange,polyval(p,xrange),'r-')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
nmc=10
xmc=createmcdata(x,xsig,nmc,'lognorm')
ymc=createmcdata(y,ysig,nmc,'lognorm')
pmc=mcpolyfitdemo(xmc, ymc, 1)
plt.errorbar(x,y,fmt='bo',xerr=xsig,yerr=ysig,capsize=5)
for i in range(nmc):
    plt.plot(xrange,polyval(pmc[i,:],xrange),'r-')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

```

```python
nmc=10000
xmc=createmcdata(x,xsig,nmc,'norm')
ymc=createmcdata(y,ysig,nmc,'norm')
pmc=mcpolyfitdemo(xmc, ymc, 1)
(pmean,psig,plow,phigh)=mcpolyfit(xmc,ymc,1,0.95)
plt.figure(1)
plt.hist(pmc[:,0], bins=50)
plt.xlabel('a')
plt.ylabel('frequency')
plt.show()
plt.figure(2)
plt.hist(pmc[:,1], bins=50)
plt.xlabel('b')
plt.ylabel('frequency')

plt.show()

```
